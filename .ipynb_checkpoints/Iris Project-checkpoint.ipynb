```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
iris=sns.load_dataset("iris")
```

```python
iris
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

iris["species"]=le.fit_transform(iris["species"])
```

```python
iris
```

```python
iris.info()
```

```python
iris.isnull().sum()
```

```python
iris.duplicated().sum()
```

```python
iris=iris.drop_duplicates(keep="first")
```

```python
iris.duplicated().sum()
```

```python
X=iris.drop("species",axis=1)
Y=iris["species"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
lg=LogisticRegression()
lg.fit(X_train,Y_train)
```

```python
y_pred=lg.predict(X_test)
```

```python
y_pred
```

```python
lg.score(X_train,Y_train)
```

```python
lg.score(X_test,Y_test)
```

```python
print("Accuracy Score:",accuracy_score(Y_test,y_pred))
```

```python
print(classification_report(Y_test,y_pred))
```

```python
print(confusion_matrix(Y_test,y_pred))
```

```python
import pickle


#pickle is a python module used to convert objects into a file and load them back later

#It saves your trained model so you dont have to train it again.
```

```python
#SAVE MODEL

with open("iris_model.pkl","wb") as f:
    pickle.dump(lg,f)
```

```python
#LOAD MODEL 

with open("iris_model.pkl","rb") as f:
    model=pickle.load(f)
```

```python
#Make new prediction


sample=[[1.5,2.5,2.0,1.9]]


prediction=model.predict(sample)
species_map={0:"Setosa",1:"Versicolor",2:"Virginica"}

print("Predicted Class:",species_map[prediction[0]])
```
